```python
# TASK 1: Creating the 3 DataFrames

import pandas as pd
import numpy as np


# 1) Employee DataFrame

employee_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005"],
    "Name": ["John Alter", "Alice Luxumberg", "Tom Sabestine", "Nina Adgra", "Amy Johny"],
    "Gender": ["M", "F", "M", "F", "F"],
    "City": ["Paris", "London", "Berlin", "Newyork", "Madrid"],
    "Age": [25, 27, 29, 31, 30]
}

df_employee = pd.DataFrame(employee_data)


# 2) Seniority Level DataFrame

seniority_data = {
    "ID": ["A001", "A002", "A003", "A004", "A005"],
    "Designation Level": [2, 2, 3, 2, 3]
}

df_seniority = pd.DataFrame(seniority_data)


# 3) Project DataFrame

project_data = {
    "ID": ["A001","A002","A003","A004","A005","A002","A005","A003","A001","A003","A001","A004","A004","A005"],
    "Project": ["Project 1","Project 2","Project 3","Project 4","Project 5","Project 6","Project 7","Project 8",
                "Project 9","Project 10","Project 11","Project 12","Project 13","Project 14"],
    "Cost": [1002000, 2000000, 4500000, 5500000, np.nan, 680000, 400000, 350000, np.nan, 300000, 2000000, 1000000, 3000000, 200000],
    "Status": ["Finished", "Ongoing", "Finished", "Ongoing", "Finished", "Failed", "Finished", "Failed",
               "Ongoing", "Finished", "Failed", "Ongoing", "Finished", "Finished"]
}



df_project = pd.DataFrame(project_data)



# Saving all 3 DataFrames as CSV

df_employee.to_csv("Employee.csv", index=False)
df_seniority.to_csv("Seniority.csv", index=False)
df_project.to_csv("Project.csv", index=False)

# Print 
df_employee, df_seniority, df_project
```

```python
# TASK 2: Replace Missing Cost Values

import pandas as pd
import numpy as np

# Loading Project CSV
df_project = pd.read_csv("Project.csv")

# create a List to store previous valid costs
previous_costs = []

# Looping through each row
for i in range(len(df_project)):
    cost = df_project.loc[i, "Cost"]
    
    # If cost is not null â†’ add to previous_costs list
    if not pd.isna(cost):
        previous_costs.append(cost)
    else:
        # running average of previous costs
        running_avg = sum(previous_costs) / len(previous_costs)
        
        # Replace missing value with running average
        df_project.loc[i, "Cost"] = running_avg
        
        # Also adding this new value to previous costs for future averaging
        previous_costs.append(running_avg)

# Saving updated CSV
df_project.to_csv("Project.csv", index=False)

df_project
```

```python
# TASK 3: Split Name into First Name and Last Name

import pandas as pd

# Loading Employee CSV
df_employee = pd.read_csv("Employee.csv")

# Split the Name column into First Name and Last Name
df_employee[['First Name', 'Last Name']] = df_employee['Name'].str.split(n=1, expand=True)

# Drop the old Name column
df_employee = df_employee.drop(columns=['Name'])

# Saving the updated Employee.csv
df_employee.to_csv("Employee.csv", index=False)

df_employee
```

```python
# TASK 4: Join all three DataFrames

import pandas as pd

# Loading all CSVs
df_employee = pd.read_csv("Employee.csv")
df_seniority = pd.read_csv("Seniority.csv")
df_project   = pd.read_csv("Project.csv")

# First join Employee + Seniority on "ID"
merged1 = pd.merge(df_employee, df_seniority, on="ID", how="left")

# Next join with Project DataFrame
Final = pd.merge(merged1, df_project, on="ID", how="left")

# final merged dataframe
Final
```

```python
# TASK 5: Add Bonus Column

import numpy as np
import pandas as pd

# Reload the Final dataframe

df_employee = pd.read_csv("Employee.csv")
df_seniority = pd.read_csv("Seniority.csv")
df_project = pd.read_csv("Project.csv")

# Merge again to get Final

merged1 = pd.merge(df_employee, df_seniority, on="ID", how="left")
Final = pd.merge(merged1, df_project, on="ID", how="left")

# Adding the new Bonus column- 5% of Cost only if the project is finished

Final["Bonus"] = np.where(Final["Status"] == "Finished", Final["Cost"] * 0.05, 0)

Final
```

```python
# TASK 6: Demote Level for Failed Projects and Remove Level > 4

import pandas as pd
import numpy as np

# Reload and recreate the Final dataframe 

df_employee = pd.read_csv("Employee.csv")

df_seniority = pd.read_csv("Seniority.csv")

df_project = pd.read_csv("Project.csv")



merged1 = pd.merge(df_employee, df_seniority, on="ID", how="left")

Final = pd.merge(merged1, df_project, on="ID", how="left")

# 1. Demote designation level by 1 if Status = Failed

Final["Designation Level"] = np.where(
    Final["Status"] == "Failed",
    Final["Designation Level"] + 1,
    Final["Designation Level"]
)

# 2. Remove all records where the designation level > 4

Final = Final[Final["Designation Level"] <= 4]

Final
```

```python

# TASK 7: Add Mr./Mrs. to First Name & Drop Gender Column

import pandas as pd
import numpy as np

# Reload data 

df_employee = pd.read_csv("Employee.csv")

df_seniority = pd.read_csv("Seniority.csv")

df_project = pd.read_csv("Project.csv")


merged1 = pd.merge(df_employee, df_seniority, on="ID", how="left")

Final = pd.merge(merged1, df_project, on="ID", how="left")


Final["Designation Level"] = np.where(
    Final["Status"] == "Failed",
    Final["Designation Level"] - 1,
    Final["Designation Level"]
)
Final = Final[Final["Designation Level"] <= 4]



# Add Mr./Mrs. based on gender

Final["First Name"] = np.where(
    Final["Gender"] == "M",
    "Mr. " + Final["First Name"],
    "Mrs. " + Final["First Name"]
)

# Dropping the Gender column
Final = Final.drop(columns=["Gender"])

Final
```

```python
# TASK 8: Promote Designation Level Based on Age

# update the 'Designation Level' column.

Final['Designation Level'] = np.where(
    # Condition: Age > 29 AND Designation Level > 1 
    (Final['Age'] > 29) & (Final['Designation Level'] > 1),
    Final['Designation Level'] - 1, # if True (Promote: Designation number decreases)
    Final['Designation Level']     # if False (Keep the current level)
)

Final
```

```python
# TASK 9: Calculate Total Project Cost per Employee

# Group the 'Final' DataFrame by the unique employee identifiers -> Calculate the sum of the Cost column ->convert the resulting Series back into a DataFrame

TotalProjCost = Final.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()

# Rename the summed 'Cost' column to 'Total Cost' 

TotalProjCost = TotalProjCost.rename(columns={'Cost': 'Total Cost'})

print("TotalProjCost DataFrame:")
TotalProjCost
```

```python
# TASK 10: Filter Employee Details by City Name

# Filter the 'Final' DataFrame.
# to find city names containing 'o' or 'O' and to ensure rows with missing City values are skipped safely.

employees_with_o_in_city = Final[Final['City'].str.contains('o', case=False, na=False)].copy()


# drop duplicate employee records

employee_details_to_print = employees_with_o_in_city[['ID', 'First Name', 'City', 'Age', 'Designation Level']].drop_duplicates()

print("Employee Details for cities containing the letter 'o':")
employee_details_to_print
```
